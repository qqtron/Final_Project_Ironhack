```python
final_df = idealista[['LOCATIONNAME','TYPOLOGY','ROOMS','AREARANGE','PRICE_ASKING']]

final_df
```

```python
area = final_df['AREARANGE']

area
```

```python
!pip install regex
import regex as re
```

```python
# removing non numerical elements to get the mean of the area range.

list_area = area.to_list()
new_list = [s.strip('m2') for s in list_area]
new_list    
```

```python
second_list = [re.sub('[\W_]+', ' ', s) for s in new_list]
area_2 = [s.strip() for s in second_list]

area_2
```

```python
area_2[0].split(" ")
```

```python
avg_area = []

for area_object in area_2:
    if ' ' in area_object:
        area_values = area_object.split()
        avg_area.append(int(int(area_values[0]) + int(area_values[1]) / 2))
    else:
        avg_area.append(int(area_values[0]))
avg_area
```

```python
final_df['AREARANGE'] = avg_area

final_df
```

```python
final_df['AREA'] = avg_area

final_df.drop(final_df['AREARANGE'])

final_df
```
