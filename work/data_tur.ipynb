```python
import pandas as pd
import numpy as np
import regex as re
```

```python
tur_data = pd.read_excel('tur.xlsx')
tur_data
```

```python
adr = list(tur_data['Avg. Daily Rate'])
print(adr[1])
adr[1].split('¬')
```

```python
clean_adr = []

for i in adr:
    try:
        x = i.split('¬')
        clean_adr.append(x[1])
    except:
        clean_adr.append(i)

#str to floats

list_to_floats = []

for item in clean_adr:
    list_to_floats.append(float(item))

list_to_floats
```

```python
tur_data['Avg. Daily Rate'] = list_to_floats

tur_data['Avg. Daily Rate']
```

```python
date = tur_data['Date']
```

```python
date[0].split(' ')
```

```python
clean_date = []

for i in date:
    try:
        x = i.split(' ')
        clean_date.append(x[1])
    except:
        clean_date.append(i)
```

```python
tur_data['Date'] = clean_date
```
